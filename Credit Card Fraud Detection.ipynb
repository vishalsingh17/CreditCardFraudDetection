1. Now here's a problem with moving forward. We cannot move forward with fitting a model on this data and training it using cross validation because that will get us biased results on the validation set and therefore eventually perform terrible on the unknown test data.
2. This is because we have undersampled the whole training data, of which the validation data would be a part of. So, while balancing the training data there would be a 'data leakage' from the validation data into training data, i.e., validation set would also be a balanced data and it will not be a representative of any real time data which is not balanced. So the model will perform not as expected on the test set as it did on the validation set
3. To tackle this issue, we would just balance the data during cross validation(CV) process, which will apply balancing only on the training folds of the CV and not on the validation fold. 
4. Since, we have already visualized the data after undersampling, this time we will rather use oversampling while training the model using CV.

```python
#we will make a funtion now to print the metrics for every estimator that we are going to be testing on
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report
from sklearn.metrics import plot_confusion_matrix

def print_metrics(accuracy_lst, precision_lst, recall_lst, f1_lst, auc_lst):
    print('Mean Accuracy: ', np.asarray(accuracy_lst).mean())
    print('Mean Precision: ', np.asarray(precision_lst).mean())
    print('Mean Recall: ', np.asarray(recall_lst).mean())
    print('Mean f1 score: ', np.asarray(f1_lst).mean())
    print('Mean AUC score: ', np.asarray(auc_lst).mean())
    plot_confusion_matrix(model, X_train.values[validation_index], y_train.values[validation_index], values_format='d')
    plt.show()
```

```python
# !pip install imblearn
```

```python
#Oversampling using SMOTE during cross validation and then training using logistic regression 
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import make_pipeline as imbalanced_make_pipeline
from sklearn.model_selection import StratifiedKFold

from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(max_iter=1000)

#applying cross validation using stratifiedKFolds- balancing and cross validation will be executed within a pipeline
def balancing_CV_pipeline(X_train, y_train, estimator):
    accuracy_lst, precision_lst, recall_lst, f1_lst, auc_lst = [],[],[],[],[]
    kf = StratifiedKFold(n_splits=5)
    for train_index, validation_index in kf.split(X_train, y_train):
        pipeline = imbalanced_make_pipeline(SMOTE(sampling_strategy='minority'), estimator)
        model = pipeline.fit(X_train[train_index], y_train[train_index])
        predictions = model.predict(X_train[validation_index])
        #now collecting metric details from our trained model and filling it in respective list after each iteration
        accuracy_lst.append(pipeline.score(X_train[validation_index], y_train[validation_index]))
        precision_lst.append(precision_score(y_train[validation_index], predictions))
        recall_lst.append(recall_score(y_train[validation_index], predictions))
        f1_lst.append(f1_score(y_train[validation_index], predictions))
        auc_lst.append(roc_auc_score(y_train[validation_index], predictions))
    return accuracy_lst, precision_lst, recall_lst, f1_lst, auc_lst, predictions, validation_index, model

(accuracy_lst, precision_lst, recall_lst, f1_lst, 
 auc_lst, predictions, validation_index, model) = balancing_CV_pipeline(X_train.values, y_train.values.ravel(), sgd_clf)
```

```python
print_metrics(accuracy_lst, precision_lst, recall_lst, f1_lst, auc_lst)
```

```python
#lets see the classification report for SGD
print(classification_report(y_train.iloc[validation_index], predictions, target_names=['No Fraud', 'Fraud']))
```

Well, we are pretty good at Recall but the Precision for detecting Fraud Transactions is extremely poor. That means we are detecting more number of False Positives or Non Fraud Transactions are Fraud, This can have a negative impact on the image and reliability of the company.

Also, from the PR curve we could indeed aim for a threshold which gives the Recall of just about 78-80% , where the Precision is also at 80%. Let's plot a graph now that plots the Precision Recall curve along with the threshold used by the decision fuction.

```python
plt.figure(figsize=(12, 6)) 
precisions, recalls, thresholds = precision_recall_curve(y_train.iloc[validation_index], y_scores[:,1])
plt.plot(thresholds, precisions[:-1], 'b--', label="Precision")
plt.plot(thresholds, recalls[:-1], 'g--', label="Recall")
plt.legend(loc="center left", fontsize=16)
plt.xlabel("Threshold", fontsize=16)
plt.grid(True)
plt.title('Plot for Precision-Recall vs Threshold', fontsize=20)
```

Now looking at the plot above, I would greatly desire that my model's threshold be around 80% to 85% to get a great set of Precision and Recall for our model. 
Let's try to achieve this.

```python
#to become more precise we can search for the lowest threshold that gives you at least 90% precision, np.argmax() will do it.
threshold_at_precision_85 = thresholds[np.argmax(precisions >= 0.86)]
#using the new threshold to update the predictions using this simple 1 line code is fun :)
new_predictions = (y_scores[:,1] >= threshold_at_precision_85).astype('int')
```
